```python
from math import pi
import matplotlib.pyplot as plt

result = pi**4 / 90
n = 75
n_digits = 20
x = [i for i in range(n)]
```

```python
def sumSeriesA(n):
    all_values = [1 / i**4 for i in range(1, n+1)]
    results = [all_values[0]]
    for i in range(1, n):
        results.append(all_values[i] + results[i-1])
    return results

def sumSeriesD(n):
    all_values = [1 / i**4 for i in range(n, 0, -1)]
    results = [all_values[0]]
    for i in range(1, n):
        results.append(all_values[i] + results[i-1])
    return results

def printDigits(n_digits, numbers):
    for i in range(len(numbers)):
        print(f'{numbers[i]:.{n_digits}f}')
```

```python
printDigits(n_digits, sumSeriesA(n))
```

```python
printDigits(n_digits, sumSeriesD(n))
```
